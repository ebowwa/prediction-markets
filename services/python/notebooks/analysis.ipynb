# Kalshi Market Analysis

This notebook connects to the Kalshi TypeScript service and analyzes market data using pandas.

```python
import sys
sys.path.append('/app/src')

from client import KalshiClient
from analysis import markets_to_dataframe, find_opportunities
import pandas as pd

# Initialize client
client = KalshiClient()
```

```python
# Fetch markets
result = client.list_markets(limit=100, status='open')
markets = result['data']['markets']

# Convert to DataFrame
df = markets_to_dataframe(markets)
df.head()
```

```python
# Find opportunities
opportunities = find_opportunities(df, min_spread=0.02)
opportunities[['title', 'yes_bid', 'yes_ask', 'spread', 'volume']].head(10)
```

```python
# Visualize price distribution
df['implied_prob'].hist(bins=50, figsize=(10, 6))
```
